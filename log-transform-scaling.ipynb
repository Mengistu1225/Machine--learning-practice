```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Values': [1, 10, 100, 1000, 10000]}
df = pd.DataFrame(data)

# Log Transformation
df['Log_Transformed'] = np.log1p(df['Values'])

print(df)

# Visualize
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.hist(df['Values'], bins=10, color='blue', alpha=0.7)
plt.title("Original Distribution")

plt.subplot(1, 2, 2)
plt.hist(df['Log_Transformed'], bins=10, color='green', alpha=0.7)
plt.title("Log Transformed Distribution")
plt.show()
```

```python

import seaborn as sns


# Create a synthetic dataset
np.random.seed(42)
data = {
    'Income': np.random.exponential(scale=50000, size=1000),  # Skewed feature
    'Expenses': np.random.exponential(scale=20000, size=1000),
    'Savings': np.random.exponential(scale=10000, size=1000)
}
df = pd.DataFrame(data)

# Check the skewness of the original features
print("Original Skewness:\n", df.skew())

# Apply log transformation
df['Income_log'] = np.log1p(df['Income'])
df['Expenses_log'] = np.log1p(df['Expenses'])
df['Savings_log'] = np.log1p(df['Savings'])

# Check the skewness of the transformed features
print("\nTransformed Skewness:\n", df[['Income_log', 'Expenses_log', 'Savings_log']].skew())

# Visualize the effect of log transformation
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
sns.histplot(df['Income'], ax=axes[0, 0], kde=True, color='blue')
axes[0, 0].set_title('Original Income')

sns.histplot(df['Expenses'], ax=axes[0, 1], kde=True, color='green')
axes[0, 1].set_title('Original Expenses')

sns.histplot(df['Savings'], ax=axes[0, 2], kde=True, color='red')
axes[0, 2].set_title('Original Savings')

sns.histplot(df['Income_log'], ax=axes[1, 0], kde=True, color='blue')
axes[1, 0].set_title('Log Transformed Income')

sns.histplot(df['Expenses_log'], ax=axes[1, 1], kde=True, color='green')
axes[1, 1].set_title('Log Transformed Expenses')

sns.histplot(df['Savings_log'], ax=axes[1, 2], kde=True, color='red')
axes[1, 2].set_title('Log Transformed Savings')

plt.tight_layout()
plt.show()
```
